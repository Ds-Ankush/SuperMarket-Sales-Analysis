## **Visulization**

```python
# QUS 81

sns.histplot(data=data , x = "Sales",bins=30, kde = True , color = "royalblue")
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
# QUS 82

branch_revenue = data.groupby("Branch")["Sales"].sum().reset_index()
sns.barplot(data = branch_revenue , x = "Sales" , y = "Branch", palette = "Set3")
plt.title("Revenue by Branch")
plt.xlabel("Revenue")
plt.ylabel("Branch")
plt.show()
```

```python
# QUS 83

city_revenue = data.groupby("City")["Sales"].sum().reset_index()
sns.barplot(data = city_revenue , x = "Sales" , y = "City", palette = "Set2")
plt.title("Revenue by City")
plt.xlabel("Revenue")
plt.ylabel("City")
plt.show()
```

```python
# QUS 84

pl_revenue = data.groupby("Product line")["Sales"].sum().reset_index()
sns.barplot(data = pl_revenue , x = "Sales" , y = "Product line", palette = "Set2")
plt.title("Revenue by product Line")
plt.xlabel("Revenue")
plt.ylabel("Product lines")
plt.show()
```

```python
# QUS 85

sns.countplot(data = data , x = "Customer type" , palette = "coolwarm")
plt.title("Customer Type Distribution")
plt.xlabel("Customer Type")
plt.ylabel("Total")
plt.show()
```

```python
# QUS 87

sns.countplot(data = data , x = "Payment" , palette = "Set1")
plt.title("Payment method Chart")
plt.xlabel("Payment Methods")
plt.ylabel("Total")
plt.show()
```

```python
# QUS 88

data["Date"] = pd.to_datetime(data["Date"])
np = (
    data.groupby(data["Date"].dt.month_name())["Sales"]
    .sum()
    .reindex(["January", "February", "March"])
    .reset_index()
)

plt.figure(figsize=(8,5))
sns.lineplot(data = np ,x= "Date", y = "Sales" , marker= "o") 
plt.title("Monthly Revenue trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

```

```python
# QUS 89

np = (
    data.groupby(data["Date"])["Sales"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(12,5))
sns.lineplot(data = np ,x= "Date", y = "Sales" , marker= "o" , color = "royalblue") 
plt.title("Daily Revenue trend")
plt.xlabel("Date")
plt.ylabel("Revenue")

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# QUS 90

plt.figure(figsize = (12,7))
sns.heatmap(data.corr(numeric_only= True) , annot = True , cmap = "coolwarm" , linewidths= 0.5)
plt.title("Correlation Heatmap")
# plt.tight_layout()
plt.show()
```
